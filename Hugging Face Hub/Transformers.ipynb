```python
import io
import sys

def get_help_text(function):
    # Cria um objeto de string para redirecionar a saída
    output = io.StringIO()

    # Salva o stdout atual
    old_stdout = sys.stdout

    try:
        # Redireciona a saída para o objeto de string
        sys.stdout = output

        # Chama a função help() com o objeto ou função desejada
        help(function)

        # Obtém o texto da ajuda
        help_text = output.getvalue()

        return help_text
    finally:
        # Restaura o stdout original
        sys.stdout = old_stdout

pipeline_help = get_help_text(pipeline)
print(pipeline_help)
```

```python
text = pipeline_help

# Converte o texto para letras minúsculas e remove pontuações
text = text.lower()
text = text.replace(".", "").replace(",", "")

words = text.split()
word_count = {}

# Contagem das palavras
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Exibe as palavras repetidas
repeated_words = [word for word, count in word_count.items() if count > 1]
print("Palavras repetidas:")
for word in repeated_words:
    print(f"{word}: {word_count[word]}")
```
