# Trabalhando com entradas não numéricas:

    Computadores não entendem letras,
    quem já ouviu falar da tabela ASCII
    sabe bem disso.

    Os caracteres estruturados numa estrutura
    de dados definida como string. Strings nada
    mais são do que vetores, porém de caracteres.

    Os caracteres em uma string são codificados
    internamente usando o conjunto de caracteres
    Unicode, que é mais abrangente do que a tabela
    ASCII. O Unicode inclui uma ampla variedade de
    caracteres de diferentes idiomas e símbolos. A
    tabela ASCII é um subconjunto do Unicode, o que
    significa que os primeiros 127 caracteres do
    Unicode correspondem diretamente à tabela ASCII.

```python
# Exemplo:
entrada = input('Digite uma frase:\n')

letras = (list(entrada))
print(letras)
print(f'Tamanho do vetor de caracteres = {len(letras)}')
```

```python
# codificando os caracteres:
vetor_numerico = [ord(letra) for letra in letras]

vetor_numerico
```

    Se fizermos o processo contrário,
    encontraremos frase!

```python
# Decodificando os caracteres:
string_caracteres = [chr(valor) for valor in vetor_numerico]
''.join(string_caracteres)
```

    Porém esse tipo de conversão, letra
    por letra, pode ser computacionalmente
    muito custoso. Então a ideia é buscar
    uma outra forma de se codificar as
    entradas não numéricas.

# Métodos de Tokenização - Modelos pré-treinados:

    Para situações ainda mais complexas,
    o hot enconding pode não ser a melhor
    opção.

    A solução para esse problema apresentada
    aqui será dada utilizando o tiktoken
    da OpenIA.

GPT (Generative Pre-trained Transformer):

    GPT é um modelo generativo que é treinado para prever a próxima
    palavra em uma sequência dada as palavras anteriores. Ele pertence
    à categoria de modelos de linguagem que geram texto continuamente,
    sendo capazes de completar frases e até mesmo criar texto coerente
    a partir de um prompt inicial.

BERT (Bidirectional Encoder Representations from Transformers):

    BERT, por outro lado, é treinado para prever palavras mascaradas
    em uma sequência. Ele utiliza uma abordagem bidirecional, considerando
    tanto as palavras anteriores quanto as posteriores para prever as
    palavras mascaradas. Isso permite que BERT capture melhor o contexto
    global de uma frase ou documento.

    Ambos GPT e BERT são exemplos de abordagens bem-sucedidas de
    pré-treinamento de modelos de linguagem que fornecem representações
    contextuais ricas, permitindo que sejam ajustados para tarefas específicas
    por meio de treinamento adicional (ajuste fino) em conjuntos de dados
    rotulados.

```python
import tiktoken
```

```python
def encode(texto):
  '''
      Função que codifica um texto,
  baseado em na tabela ASCII.
  '''

  letras = (list(texto))

  vetor_numerico = [ord(letra) for letra in letras]

  return vetor_numerico
```

```python
enc = tiktoken.get_encoding("gpt2")
enc.encode(texto)
```

```python
# Comparando o tamanho das respostas:
print(f'Tamnho da mensagem codificada -> {len(encode(texto))}')
print(f'Tamanho da codificação retrabalhada pelo GPT -> {len(enc.encode(texto))}')
```

**Resumindo**

    A ideia aqui é utilizar o GPT para
    retrabalhar o encoding, de modo
    que se consiga, mesmo que codificando
    letra por letra, se reduzir o tamanho
    do vetor resposta sem perder a informação.

    Isso é muito importante para conseguir
    ter um bom modelo de LLM sem ficar
    restrito pelo poder de processamento.

```python
PATH = './sample_data/'
dados_treino = 'crepusculoDosIdolos.txt'
```

```python
from tokenizers import ByteLevelBPETokenizer
```

    O ByteLevelBPETokenizer faz uma recursividade contando
    as frequências de pares de tokens, ou seja, ele utiliza
    contagem de frequência para filtrar a contagem de tokens.

    Isso é bem util para a questão aliviar o custo computacional
    desse tipo de modelo.

```python
# Instanciando o modelo:
tokenizer = ByteLevelBPETokenizer()

tokenizer.train(files=[PATH+dados_treino],vocab_size=52_000,
                min_frequency=2, special_tokens=[
                    "<s>",
                    "<pad>",
                    "</s>",
                    "<unk>",
                    "<mask>"])
```

vocab_size

    Parâmetro que define o tamanho de espaços para armazenar
    todas as possibilidades, ou seja, é quem define a complexidade
    do vocabulario.

min_frequency

    Parâmetro que define o treshold(limite minimo) de ocorrências
    dos tokens para que estes sejam "juntados" (somados, no caso).

## Processamento:

```python
# Criando o tonkenizer baseado no algoritmo BPE
from tokenizers.implementations import ByteLevelBPETokenizer
from tokenizers.processors import BertProcessing
```

```python
# Instanciando modelo:
tokenizer = ByteLevelBPETokenizer(
    PATH+'RAW_MODEL'+'/vocab.json',
    PATH+'RAW_MODEL'+'/merges.txt')

tokenizer._tokenizer.post_processor = BertProcessing(
    ("</s",tokenizer.token_to_id("</s>")),
    ("<s>", tokenizer.token_to_id("<s>"))
)
tokenizer.enable_truncation(max_length=512)
```

    Aqui vamos utilizar a Roberta, que é o
    modelo BERT do google, otimizado pelo
    pessoal da META.

    A Roberta é um encoder baseado em transformers
    otimizado para máquinas de menor recurso
    e para menores quantidade de dados.
  


Construindo nosso Tokenizer

```python
# Imports
from transformers import RobertaTokenizer
from transformers import RobertaForMaskedLM
```

```python
# Instanciando o Tokenizer do LLM escolhido:
tokenizer = RobertaTokenizer.from_pretrained(PATH+'RAW_MODEL',max_len=512)
```

Criando nosso Transformer

```python
from transformers import RobertaConfig
```

```python
 # Configurando o Transformer
 config = RobertaConfig(
     vocab_size=52_000,
     max_position_embeddings = 512,
     num_attention_heads = 12,
     num_hidden_layers = 6,
     type_vocab_size = 1)
```

```python
# Instanciando Modelo:
model = RobertaForMaskedLM(config=config)
```

```python
# Quantidade de Parâmetros da rede Neural:
model.num_parameters()
```

Criando nosso Dataset tokenizado

```python
from transformers import LineByLineTextDataset
```

```python
# Tokenizando os dados de Treino:
dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path=PATH+dados_treino,
    block_size=128
)
```

```python
# Verificando
dataset.examples[:2]
```

```python
tokenizer.decode(dataset.examples[7]['input_ids'])
```

Treinando o Modelo:

```python
from transformers import DataCollatorForLanguageModeling
```

```python
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.1)
```

mlm

    Indica se terá mascaramento.

mlm_probability

    A probabilidade do mascaramento

Nota:

    O codecarbon é uma API focada em calcular o gasto energético
    de uma aplicação, no caso, o gasto energético do treinamento
    do nosso modelo.

    Eu achei algo interessante de se experimentar, inclusive para
    ter mais uma medida de desempenho do modelo.

```python
from transformers import Trainer, TrainingArguments
from codecarbon import EmissionsTracker
```

```python
# Elementos do treinamento:
training_args = TrainingArguments(
    output_dir=PATH+'RAW_MODEL',
    overwrite_output_dir=True,
    num_train_epochs=1200,
    per_device_train_batch_size=64,
    save_steps=10_000,
    save_total_limit=2,
    prediction_loss_only=True)

# Treinanmento:
Trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset)

# Instanciando Tracker de Emissões de Carbono:
tracker = EmissionsTracker()
```

```python
# Iniciando:
tracker.start()

# Fit:
Trainer.train()
```

```python
# Interrompendo:
tracker.stop()
```

```python
Trainer.save_model(PATH+'RAW_MODEL')
```
