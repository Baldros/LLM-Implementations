```python
# Contar a frequência dos valores na coluna 'owned_by'
counts = Models['owned_by'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
counts.plot(kind='barh', color='skyblue')

# Adicionar título e rótulos aos eixos
plt.title('Contagem de Modelos por Proprietário')
plt.xlabel('Proprietário')
plt.ylabel('Contagens')

# Exibir o gráfico
plt.show()
```
